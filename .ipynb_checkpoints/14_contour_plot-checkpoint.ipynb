# 3D Area Plotting

Dalam sesi ini kita akan mempelajari cara membuat 3D area plotting dengan Matplotlib.

## Import modules

```python
# %matplotlib inline
%matplotlib notebook 
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Sample dataset

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T # transpose
z = np.cos(x ** 2 + y ** 2)

print(x.shape)
print(y.shape)
print(z.shape)
```

```python
z
```

## 3D Wireframe

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='green')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')
plt.show()
```

## 3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, color='green')
# ax.plot_surface(x, y, z, cmap='viridis')


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')
plt.show()
```
