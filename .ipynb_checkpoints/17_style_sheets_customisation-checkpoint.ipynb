# Matplotlib Customisation: Style Sheets

Dalam sesi ini kita akan mempelajari kustomisasi Matplotlib dengan Style Sheets.

## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Predefined Styles

Matyplotlib menyediakan beberapa predefined styles. Detail dari tiap predefined style bisa kita akses di [sini](https://github.com/matplotlib/matplotlib/tree/master/lib/matplotlib/mpl-data/stylelib) 

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 25)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

## Membuat Style sendiri
